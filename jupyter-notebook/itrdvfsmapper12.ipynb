```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
```

```python
df_comb = pd.read_csv("../logs/5_11_2024_itrdvfsfor12mappers//combined.csv")
```

```python
df_comb = df_comb[(df_comb['SourcenumRecordsOutPerSecond_avg'] / df_comb['rate']  > 0.90)]
df_comb
```

```python
df_comb.plot(x="rate", y=["SourcenumRecordsOutPerSecond_avg"], kind="line") 
```

```python
print(df_comb['rate'].unique())
for qps in df_comb['rate'].unique():
    df_comb[(df_comb.itr == 2) & (df_comb.rate==qps)].plot(x='dvfs', y='pkg_watts_avg', ylim=[0, 120], kind='scatter', title=f"rate={qps}")
```
