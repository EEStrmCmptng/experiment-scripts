```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import matplotlib.patches as mpatches
from matplotlib.markers import MarkerStyle
from matplotlib.colors import Normalize


```

```python
df_comb = pd.read_csv("../logs/qps100000//combined.csv")
```

```python
df_comb
```

```python
print(df_comb.columns)
```

```python
dfondemand = df_comb[df_comb.policy == 'ondemand']
dfondemand
```

```python
dfuserspace = df_comb[df_comb.policy == 'userspace']
dfuserspace
```

```python
ax = dfuserspace.plot(x="SourcebackPressuredTime_%", y=["pkg_watts_avg"], kind="scatter", color='red', label='userspace') 
dfondemand.plot(ax=ax, x="SourcebackPressuredTime_%", y=["pkg_watts_avg"], kind="scatter", color='blue', label='ondemand') 
```

```python
plt.rcParams['figure.figsize'] = 25, 10
plt.rc('axes', labelsize=30)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=30)    # fontsize of the tick labels
plt.rc('ytick', labelsize=30)    # fontsize of the tick labels

fig, ax = plt.subplots()
# Scatter plot for C1
scatter_c1 = ax.scatter(dfuserspace['SourcebackPressuredTime_%'].tolist(),
                        dfuserspace['pkg_watts_avg'].tolist(), s=300, c=dfuserspace.itr, edgecolor="black", cmap='Blues', marker=MarkerStyle("o"))
# Add colorbars
cbar_c1 = plt.colorbar(scatter_c1, ax=ax, orientation='vertical', label="ITR-delay (us)")
ax.set_xlabel('SourceBackPressure%')
ax.set_ylabel('Power')
    
ax.set_title(f"Query1 @ 100K RPS", size=14)
plt.tight_layout()
```
